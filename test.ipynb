```python
import numpy as np
from keras.models import load_model
from keras.utils import CustomObjectScope
import tensorflow as tf
```

```python
with CustomObjectScope({'tf':tf}):
    model = load_model("./nn4.small2.v1.h5")
```

```python
model.summary()
```

```python
from skimage.io import imread
```

```python
imgs = []
for i in range(1,4):
    imgs.append(imread("img%d.jpg"%i)/255.)
```

```python
imgs = np.array(imgs)
imgs.shape
```

```python
embs = model.predict(imgs)
embs.shape
```

```python
org = embs[0]
cod = embs[1:]
org.shape, cod.shape
```

```python
def cosine_dist(org, cod):
    return cod.dot(org)/(np.linalg.norm(org)*np.linalg.norm(cod, axis=-1))
cosine_dist(org, cod)
```

```python
def euclidian_dist(org, cod):
    return np.linalg.norm(cod-org, axis=-1)
euclidian_dist(org, cod)
```
